```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_0 = pd.read_csv('Data_frames/TEMP_CH.csv', encoding = 'Latin-1')
```

```python
df_0.head()
```

```python
df_1 = df_0[df_0.Element == 'Temperature change']
```

```python
df_1.head()
```

```python
df_2 = df_1.drop(['Area Code','Months Code', 'Element Code', 'Element', 'Unit'], axis = 1)
```

```python
df_2.head()
```

```python
df_Italy = df_2[df_2.Area == 'Italy']
```

```python
df_Italy
```

```python
df_France = df_2[df_2.Area == 'France']
```

```python
df_France
```

```python
df_Mexico = df_2[df_2.Area == 'Mexico']
```

```python
df_Mexico
```

```python
df_Japan = df_2[df_2.Area == 'Japan']
```

```python
df_Japan
```

```python
x_year = np.array([x for x in range(1961,2020)])
```

```python
y_It_year = df_Italy.iloc[16,2:61]
y_Fr_year = df_France.iloc[16,2:61]
y_Me_year = df_Mexico.iloc[16,2:61]
y_Jp_year = df_Japan.iloc[16,2:61]
```

```python
plt.figure(figsize = (12,20))

plt.subplot(4,1,1)
plt.plot(x_year,y_It_year)
plt.title("Annual Temperature change from 1960 to 2019", fontsize = 16)
plt.text(1965,1.5, str("Temperature change in Italy"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,2)
plt.plot(x_year,y_Fr_year)
plt.text(1965,1.5, str("Temperature change in France"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,3)
plt.plot(x_year, y_Me_year)
plt.text(1965,1.2, str("Temperature change in Mexico"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,4)
plt.plot(x_year, y_Jp_year)
plt.text(1965,1.0, str("Temperature change in Japan"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.tight_layout()
```

```python
y_It_w = df_Italy.iloc[12,2:61]
y_It_s = df_Italy.iloc[14,2:61]
y_It_f = df_Italy.iloc[15,2:61]
y_It_sp = df_Italy.iloc[13,2:61]
```

```python
plt.figure(figsize = (12,20))

plt.subplot(4,1,1)
plt.plot(x_year,y_It_f)
plt.title("Annual Temperature change for seasons in Italy from 1960 to 2019", fontsize = 16)
plt.text(1965,1.5, str("Temperature change in Autumn"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,2)
plt.plot(x_year,y_It_w)
plt.text(1965,1.5, str("Temperature change in Winter"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,3)
plt.plot(x_year, y_It_sp)
plt.text(1965,1.7, str("Temperature change in Spring"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.subplot(4,1,4)
plt.plot(x_year, y_It_s)
plt.text(1965,3.8, str("Temperature change in Summer"), fontsize = 14, color = 'b')
plt.ylabel("Temperature change (°C)", fontsize = 12)


plt.tight_layout()
```
